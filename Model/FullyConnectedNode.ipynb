```python
import folium
import pandas as pd
```

```python
df_locations = pd.read_csv('./Data/15loc.csv')
df_locations
```

```python
# Initialize map
map_center = [df_locations['lat'].mean(), df_locations['lon'].mean()]
custom ="cartodb positron"
map = folium.Map(location=map_center, zoom_start=4, tiles=custom)

# Choose location
locations = df_locations

# Name of map
map_name = "./maps/FullyConnectedNode.html"
map.save(map_name)
```

```python
# Add marker
for _, row in df_locations.iterrows():
    folium.Circle(
        location=[row['lat'],row['lon']],
        radius=10000,  # Radius in pixels
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.6,
        tooltip=row['name'],
    ).add_to(map)
        

map.save(map_name)
```

```python
# Funtinon for find coordinates of all Connected node
def ConnectedAllNode(df):
    coordinates = []
    n = len(df)
    for i in range(len(df)):
        for j in range(i+1, len(df)):
            coordinate = []
            point1 = [df.iloc[i]['lat'], df.iloc[i]['lon']]
            point2 = [df.iloc[j]['lat'], df.iloc[j]['lon']]
            coordinate.append(point1)
            coordinate.append(point2)
            coordinates.append(coordinate)
    return coordinates
```

```python
coordinates = ConnectedAllNode(df_locations)
# Add edge for all connected node
folium.PolyLine(
    locations=coordinates,
    color="green",
    weight=0.2,
    tooltip="All connected node"
).add_to(map)

map.save(map_name)
```
